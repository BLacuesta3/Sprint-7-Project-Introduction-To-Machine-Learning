### Exploratory Data Analysis and Data Checking/ Cleaning

```python
#Import the necessary libraries.
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import accuracy_score, recall_score

from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, classification_report

from sklearn.model_selection import GridSearchCV
```

```python
#Create the dataframe.  Use the try and except clause in order to make sure that the dataframe will upload on a local and non local computer.
try:
    df = pd.read_csv('users_behavior.csv')
except:
    df = pd.read_csv('datasets/users_behavior.csv')

df.head()
```

### Splitting the Data

```python
#Create a two dataframes, one that contains the features and one that contains the target.
features = df.drop(['is_ultra'], axis=1)

target = df['is_ultra']
```

```python
#Use the train_test_split() method in order to split the data to a 60/20/20 ratio.
features_train, features_valid_test, target_train, target_valid_test = train_test_split(features, target, test_size=0.4, random_state=135)

features_valid, features_test, target_valid, target_test = train_test_split(features_valid_test, target_valid_test, test_size=0.5, random_state=246)
```

# Logistic Regression

### Logistic Regression GridSearch

```python
#Create a dictionary of parameters with lists of possible parameters for GridSearch.
lr_param_grid = {'penalty': ['l1', 'l2'],
                           'C': [0.01, 0.1, 1, 10, 100],
                           'solver': ['liblinear', 'saga']
    
}
```

```python
#Create a Logistic Regression model using the LogisiticRegression() function from the linear_model  module of the sklearn library.
model = LogisticRegression(random_state=54321)
```

```python
#Use the GridSearchCV() function in order to create a Gridsearch model in order to choose the best set of parameters for the Logistic Regression model.
lr_gridsearch = GridSearchCV(estimator=model, param_grid=lr_param_grid, cv=5, scoring='accuracy')

#Use the fit() method to train the gridsearch model.
lr_gridsearch.fit(features_train, target_train)

#Print out the best parameters and the best score for the gridsearch model.
print("Logistic Regression Best Parameters:", lr_gridsearch.best_params_)
print("Logisitc Regression Best Score:", lr_gridsearch.best_score_)
```

```python
#Enter the parameters from GridSearch into the Logistic Regression model.
model = LogisticRegression(C=10, penalty='l1', solver='liblinear')

#Train the Logistic Regression model using the fit() method.
model.fit(features_train, target_train)
```

```python
#Use the score() method in order to calculate the model's accuracy score when the model uses the training set as input.
score_train = model.score(features_train, target_train)

#Use the score() method in order to calculate the model's accuracy score when the model uses the validation set as input.
score_valid = model.score(features_valid, target_valid)
```

### Logistic Regression Training Set Accuracy Score

```python
#Print the accuracy score of the model's performance on the training set.
print("Logistic Regression Training Set Accuracy Score:", score_train)
```

### Logistic Regression Validation Set Accuracy Score

```python
#Print the accuracy score of the model's performance on the training set.
print("Logistic Regression Validation Set Accuracy Score:", score_valid)
```

Conclusion:
The logistic regression model has an accuracy score of: 0.752721617418385 when performing tasks on the training set.  This accuracy score is greater than the threshold of 0.75 accuracy for this project.  This accuracy score is satisfactory.

### Logistic Regression Prediction Validation Model 

```python
#Use the predict method on the model with the features from the validation set in order to have the model predict the target of a new set of data. 
#This prediction will be used for the confusion matrix. 
prediction_1 = model.predict(features_valid)
```

### Confusion Matrix Logistic Regression

```python
#Use the confusion_matrix() fucntion from the metrics module of the sklearn library in order to error test the model using the confusion matrix algorithm.
conf_matrix_1 = confusion_matrix(target_valid, prediction_1)

#Print the confusion matrix of the Logistic Regression Model.
print(f"Logisitic Regression Confusion Matrix:\n{conf_matrix_1}")
```

# Decision Tree

### Decision Tree Gridsearch

```python
#Create a dictionary that contains lists of possible parameters for the Decision Tree model.
tree_param_grid ={
    'max_depth' : [2, 4, 6, 8, 10, 12, 15],
    'min_samples_split': [2, 5, 10, 20],
    'max_features': ['auto', 'sqrt', 'log2'],
    'criterion': ['gini', 'entropy']}
```

```python
#Create a Decision Tree model using the DecisionTreeClassifier() function from the tree module of the sklearn library.
model_2 = DecisionTreeClassifier(random_state=123)
```

```python
#Use the GridsearchCV() function in order to find the set of best parameters for the Decision Tree model.
tree_grid_search = GridSearchCV(estimator=model_2, param_grid=tree_param_grid, cv=5, scoring='accuracy')

#Use the fit() method in order to train the GridSearch model.
tree_grid_search.fit(features_train, target_train)

#Print the best set of parameteres and the best score for the Decision Tree model.
print("Decision Tree Best Parameters:", tree_grid_search.best_params_)
print("Decison Tree Best Score:", tree_grid_search.best_score_)
```
